```
import pandas as pd
import matplotlib.pyplot as plt
```

 ##Add the total score column and Percentage column

```
df = pd.read_excel("drive/MyDrive/StudentsPerformance.xlsx")
df["total_score"] = df["math score"] + df["reading score"] + df["writing score"]
df["percentage"] = (df["total_score"] / 300) * 100
df
```

```
df
```

##To add the Examiners Comment

```
max = df[df.percentage == 100]
max
df["comment"] = "good"
new_df = df
```

```
new_df.loc[df['percentage'] >= 70, 'comment'] = 'Excellent'
new_df.loc[(df['percentage'] < 70) & (df["percentage"] <= 60),'comment' ]= 'Very Good'
new_df.loc[(df['percentage'] < 60) & (df["percentage"] >= 50),'comment' ] = 'Good'
new_df.loc[df['percentage'] < 50, 'comment'] = 'Fail'
new_df.sort_values(by=["percentage"], ascending=False)
new_df.to_excel("new_studentsPerformance.xlsx")
```

##Find the Overall Best student

```
best_overall_student = df[df.percentage == df["percentage"].max()]
best_overall_student
```

There are three students that scored the highest mark

##Find the Worst Student

```
 worst_overall_student = df[df.percentage == df["percentage"].min()]
worst_overall_student
```

##To rearrange using race/ethnicity

```
new_df.loc[df["race/ethnicity"] == "group A", "race/ethnicity"] = "A"
new_df.loc[df["race/ethnicity"] == "group B", "race/ethnicity"] = "B"
new_df.loc[df["race/ethnicity"] == "group C", "race/ethnicity"] = "C"
new_df.loc[df["race/ethnicity"] == "group D", "race/ethnicity"] = "D"
new_df.loc[df["race/ethnicity"] == "group E", "race/ethnicity"] = "E"
standard = new_df.sort_values(by = ["race/ethnicity"], ascending = True)
standard.to_csv("SP.csv", index=F)
```
